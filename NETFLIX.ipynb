#ANÁLISE EXPLORATÓRIA


```python
#identificando quantos títulos possuem em 'Movie" e em 'TV show'
types =df.groupby(['type',])['type'].count().reset_index(name='count')
types = types.set_index('type')
types
```

Aqui podemos concluir que temos 6126 filmes e 2664 programas de TV.


```python
#Agrupando diretores, tipo de progarma, e quantos 'shows'
show_director = df.groupby(['director','type'])['director'].count().reset_index(name= 'show_count')
show_director
```

```python
#top 10 melhores diretores
top10 = show_director.query(" show_count >=12")
top10
```

Chegamos a conclusão dos 10 diretores com maior qunantidade de obras da netflix. "Not Given" são dados como não específicado.
